```python
import numpy as np

data = np.random.randn(1000, 5)
z = (data - data.mean(axis=0)) / data.std(axis=0)

# Flatten and find top 3
flat = z.flatten()
top3_idx = flat.argsort()[-3:][::-1]
top3_values = flat[top3_idx]
top3_positions = [np.unravel_index(i, z.shape) for i in top3_idx]

print("Top 3 Values:", top3_values)
print("Positions:", top3_positions)
```
