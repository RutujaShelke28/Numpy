```python
import numpy as np
import time

l1 = [1, 2, 3]
l2 = [3, 4, 5]

list(zip(l1,l2))

##speedtest using python list
size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x, y in zip(l1,l2)]
end = time.time()
print(end - start)
print(add[0:10])

##speedtest using numpy arrays
size = 1_000_000

arr1 =np.array( list(range(size)))
arr2 = np.array(list(range(size)))

start = time.time()
add = arr1 + arr2
end = time.time()
print(end - start)
print(add[0:10])
```
